```python
import pickle as pkl
import seaborn as sns
```

```python
input_path = '/usr/local/google/home/fprost/Downloads/biosbias/CC-MAIN-2018-34-bios.pkl'
```

```python
data = pkl.load(open(input_path, 'rb'))
```

```python
data[0]
```

```python
import matplotlib.pyplot as plt
```

# Gender

```python
plt.rcParams['figure.figsize'] = [5, 5]
sns.countplot([x['gender'] for x in data])
```

# Professions

```python
plt.rcParams['figure.figsize'] = [15, 10]
plt.xticks(rotation=90)
sns.countplot([x['title'] for x in data])
```

# Professions & Gender

```python
plt.rcParams['figure.figsize'] = [15, 10]
plt.xticks(rotation=90)
sns.countplot(x=[x['title'] for x in data], hue=[x['gender'] for x in data])
```
